```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
#using MATLAB

#Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, label = 'Line')
plt.legend()
plt.show()
```

```python
#Using MATLAB with Object-Oriented

#Creating a figure
fig, ax = plt.subplots()
#Define the axes
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#------------------------------------Matlabplot------------------------------

# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker = "*", color = "red", linewidth=2 ,label = 'Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0,45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
plt.show()
```

```python
#------------------------------------Matlabplot Orienteded Objetc------------------------------

# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis, marker = "D", linewidth=2 ,color = "green", label = "Boston")
ax.set_xlabel("Date")
ax.set_ylabel("fare($)")
ax.set_title("PyBer Fare by Month")
ax.legend()
ax.grid()
```

```python
#-----------------------------------------BAR CHART------------------------------------

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.bar(x_axis, y_axis, color ="green", label = "Boston")
# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis, color = "Magenta", label = "Boston")
# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.gca().invert_yaxis()
plt.legend()
```

```python
#------------------------------------BAR CHART Orienteded Object------------------------------------
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="black", label='New York')
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')
ax.set_xlabel("Date")
ax.set_ylabel("fare($)")
ax.set_title("PyBer Fare by Month")
```

```python
#------------------------------Scatter Plot-------------------------------
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis,x_axis, c = "r", alpha=0.5, label = "Chicago")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.gca().invert_yaxis()
plt.xlim(0,50)
plt.legend()
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s= y_axis_larger)
```

```python
plt.scatter(y_axis,x_axis, c = "#14213d", label = "Boston", alpha=0.2, edgecolors="#001219", linewidth=2, s = [i * 5 for i in y_axis])
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("ByBer Fare by Month")
plt.gca().invert_yaxis()
plt.xlim(0,50)
plt.legend()
```

```python
#-----------------------------------PIE CHART--------------------------------
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", 
          "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, labels = x_axis, autopct='%.1f%%', colors = colors)
plt.show()
```

```python
#------------with object oriented-------------
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)

plt.show()
```
